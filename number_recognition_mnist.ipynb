```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
#reading the file by providing its path
data = pd.read_csv('C:/Users/USER/Desktop/py/mc_ai/train.csv')
```

```python
data
```

```python
#extracting data from the dataset and viewing them up close
a = data.iloc[3,1:].values
```

```python
#reahaping the extracted data into a reasonable size
a = a.reshape(28,28).astype('uint8')
plt.imshow(a)
```

```python
#seperating labels from the data valuse as it is not necessary
#preparing the data
df_x = data.iloc[:,1:]
df_y = data.iloc[:,0]
```

```python
#creating test and train sizes or batches
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size = 0.2, random_state=4)
```

```python
#checking the data
y_train.head()
```

```python
x_train.head()
```

```python
#calling the classifier
rf = RandomForestClassifier(n_estimators=100)
```

```python
#fitting the model
rf.fit(x_train, y_train)
```

```python
#prediction on test data
pred = rf.predict(x_test)
```

```python
#the program outputs an array based on its training
pred
```

```python
#comparing prediction with actual data
s = y_test.values

#to calculate number of correctly predicted values
count = 0 
for i in range(len(pred)):
    if pred[i] == s[i]:
        count = count+1
```
